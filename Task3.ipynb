# Boxplot

```python
df.boxplot(column=["Profit"])
```

```python
print(df["Profit"].skew())
```

```python
print(df["Profit"].quantile(0.10))
print(df["Profit"].quantile(0.90))
```

```python
df["Profit"]=np.where(df["Profit"]<-15.47346,-15.47346,df["Profit"])
df["Profit"]=np.where(df["Profit"]>89.28162,89.28162,df["Profit"])
print(df["Profit"].skew())
```

```python
df["Sales"].describe()
```

# Heatmap

```python
plt.figure(figsize=(10,7))
sns.heatmap(df.corr(),annot=True,xticklabels="auto",yticklabels="auto",linewidth=2.5,cmap="YlGnBu")
```

```python
a=pd.pivot_table(df,index="Segment",values="Profit")
a.plot(kind="bar",subplots=True,figsize=(10,5),color=['skyblue'])
```

```python
b=pd.pivot_table(df,index="Sub-Category",values="Profit")
b.plot(kind="bar",color="lightGreen",figsize=(10,5),fontsize=10)
```

```python
c=pd.pivot_table(df,index="State",values="Sales")
c.plot(kind="bar",color="lightgray",edgecolor="darkblue",figsize=(20,10),fontsize=15)
```

# Barplot

```python
sns.barplot(data=df,x="Region",y="Profit",hue="Segment",palette="Paired")
plt.show()
```

```python
sns.barplot(data=df,x="Region",y="Profit",hue="Category",palette="tab20")
plt.show()
```

# Linegraph

```python
plt.figure(figsize=(10,4))
sns.lineplot(x="Quantity",y="Profit",data=df,color="Cyan",label="Profit Vs Quantity")
plt.legend()
```

```python
plt.figure(figsize=(10,4))
sns.lineplot(x="Discount",y="Profit",data=df,color="firebrick",label="Profit Vs Discount")
plt.legend()
```

1. Superstore obtain loses by giving discounts 
2. but if giving discounts if stopped then there will be an impact on the sales and may may result in attracting customers. 
3. The small loses from the discounts are essential part of company's business.
